```python
# Standardization of Numericals using z-score
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].head(5)
```

```python
# Making a copy of the numericals
scaleddf2=df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].copy()
```

```python
scaleddf2
```

```python
# Now apply the formula 
# zscore= value-mean/standard deviation
for i in ['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']:
    scaleddf2[i]=(scaleddf2[i]-scaleddf2[i].mean())/scaleddf2[i].std()
```

```python
# These are the standardized set of numerical values
# all the values  below -3 and above 3 are outliers
# this is also a way of detecting outliers
scaleddf2
```

```python
# BEFORE standardization:
sns.displot(df['Volume'])
```

```python
# AFTER Standardization:
sns.displot(scaleddf2['Volume'])
```

```python
#Conversion of categorical values to dummy variables using one-hot encoding
```

```python
catdf=df[['Day','City','BU','Brand','Model']].copy()
catdf.sample(5)
```

```python
ohe=pd.get_dummies(df.Day,dtype=int)
```

```python
ohe.sample(5)
```

```python
# Now we merge it with our catdf and remove the 'Day' column
catdf=pd.concat([catdf,ohe],axis='columns')
```

```python
#we follow the same for rest 
ohe1=pd.get_dummies(df.BU,dtype=int)
catdf=pd.concat([catdf,ohe1],axis='columns')
```

```python
ohe2=pd.get_dummies(df.Brand,dtype=int)
catdf=pd.concat([catdf,ohe2],axis='columns')
```

```python
ohe3=pd.get_dummies(df.Model,dtype=int)
catdf=pd.concat([catdf,ohe3],axis='columns')
```

```python
catdf
```

```python
# therefore this is the final dataframe where all the categorical data is one-hot encoded
# this form of data is easy to feed into ml models to get good predictions and higher accuracy
catdf=catdf.drop(['Day','City','BU','Brand','Model'],axis='columns')
catdf
```

```python
finalcatdf=catdf.copy()
```
