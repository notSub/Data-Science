```python
#importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.Data Cleaning and Preparation:
```

```python
#Loading the csv file into a dataframe
df=pd.read_csv('Cardiotocographic.csv')
df
```

```python
#Tells us the number of rows and columns
df.shape
```

```python
# this shows the datatype of each variable and they all seem to be of the datatype float
# The variables LB,AC,DS,DP,MLTV,WIDTH,TENDENCY,NSP all have null values
df.info()
```

```python
# This command helps us return a dataframe of rows with the null values
null = df[df.isnull().any(axis=1)]
```

```python
null.shape
```

```python
# Here we can see that the missing values are only in these 21 rows out of 2126 rows
# since 1% of the values are only missing we can just delete them rather than impute them
null.sample(5)
```

```python
# Dropping all the rows that have missing values
newdf=df.dropna()
newdf
```

```python
# Now we can see that there are no more missing values.
# Dtype is also of a consistent form
# Thus cleaning of the data is done
newdf.info()
```

```python
# 2.Statistical Summary:
```

```python
# This describes all the measures of central tendency for each variable
# 1.LB -  Fetal heart rate over a period of time 
# 2.AC - Could represent "Accelerations" in the FHR. Accelerations are usually a sign of fetal well-being.
# 3.FM - May indicate "Fetal Movements" detected by the monitor.
# 4.UC - Likely denotes "Uterine Contractions", which can impact the FHR pattern.
# 5.DL - Could stand for "Decelerations Late" with respect to uterine contractions, which can be a sign of fetal distress.
# 6.DS - May represent "Decelerations Short" or decelerations of brief duration.
# 7.DP - Could indicate "Decelerations Prolonged", or long-lasting decelerations.
# 8.ASTV - Might refer to "Percentage of Time with Abnormal Short Term Variability" in the FHR.
# 9.MSTV - Likely stands for "Mean Value of Short Term Variability" in the FHR.
# 10.ALTV - Could represent "Percentage of Time with Abnormal Long Term Variability" in the FHR.
# 11.MLTV - Might indicate "Mean Value of Long Term Variability" in the FHR.

# HIGHLIGHTS
# - Mean is 133.3 which seems to be the average fetal heart rate and hence this variable is normal
# - All the varibales Except the LB variable have a negative value as minimum. This is ambiguous because only the descriptions of DL,DS,DP variables 
# suggest they might have a negative value. All the other variable must be having a positive value
# All the variables are numerical variables and there are no categorical variables.

newdf.describe()
```

```python
# 3.Data Visualization:
```

```python
# Histogram for the variable LB
# LB variable seems to have a normal distribution with outliers at the extremes with fetal hear rate of 210 and 50
# This means that the fetal heart rate is within the range of 120 to 160 which seems to be the average and normal
plt.hist(newdf['LB'],edgecolor='black',bins=30)
plt.xlabel('Fetal Heart Rate')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for the variable AC
# There are negative values in the AC which is ambiguous as accelerations cant be negative unless its deceleration
# And all these negative values seem to be outliers
# Follows positively skewed distribution
plt.hist(newdf['AC'],edgecolor='black',bins=30)
plt.xlabel('Accelerations in FHR')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for the variable FM
# This shows the plot of fetal movements and the value ranges between -0.1 to 0.2 which means that there are no movements basically
# The negative value also seems abnormal for this variable and there are outliers beyond +-0.2
# again a postively skewed distribution
plt.hist(newdf['FM'],edgecolor='black',bins=30)
plt.xlabel('Fetal Movements')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for the variable UC
# Again negative values present which is ambigous as contractions cant be negative
# Seems to follow normal distribution
plt.hist(newdf['UC'],edgecolor='black',bins=30)
plt.xlabel('Uterine Contractions')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot for the variable DL- Deceleration Late
# Mean is close to 0
# Anything above 0.1 and below 0 are outliers
plt.boxplot(newdf['DL'])
plt.show()
```

```python
# Boxplot for the variable DS- Deceleration Short
# Mean is close to 0
plt.boxplot(newdf['DS'])
plt.show()
```

```python
# Boxplot for the variable DP- Deceleration Prolonged
# Mean is close to 0
plt.boxplot(newdf['DP'])
plt.show()
```

```python
# Boxplot for the variable ASTV-Abnormal Short Term Variability
# Mean is around 50
# anything beyond 100 and below 10 are outliers
plt.boxplot(newdf['ASTV'])
plt.show()
```

```python
# Violinplot for the variable MSTV-Mean Value of Short Term Variability
# Mean is around 1
sns.violinplot(newdf['MSTV'])
```

```python
# Violinplot for the variable ALTV -Percentage of Time with Abnormal Long Term Variability
# Mean is around 0
sns.violinplot(newdf['ALTV'])
```

```python
# 4.Pattern Recognition and Insights:
```

```python
# Bi-Variate Analysis
# We use the correlation matrix to find any association
newdf.corr()
```

```python
#From give data we going to test for: from the corr matrix and given description of variables
# LB vs AC- Accelerations are usually a sign of fetal well-being.
# LB vs UC- UC which can impact the FHR pattern
# UC vs DL- which can be a sign of fetal distress.
# LB vs MLTV- Might indicate "Mean Value of Long Term Variability" in the FHR.
```

```python
# LB vs AC- Accelerations are usually a sign of fetal well-being.
# From the Graph no credible insight can be drawn
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(newdf['LB'], newdf['AC'])
ax.set_xlabel('(Fetal Heart rate)')
ax.set_ylabel('(Accelerations)')
plt.show()
```

```python
# LB vs UC- UC which can impact the FHR pattern
# From the Graph no credible insight can be drawn

fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(newdf['LB'], newdf['UC'])
ax.set_xlabel('(Fetal Heart rate)')
ax.set_ylabel('(Uterine Contractions)')
plt.show()
```

```python
# UC vs DL- which can be a sign of fetal distress.
# From the Graph no credible insight can be drawn
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(newdf['UC'], newdf['DL'])
ax.set_xlabel('(Uterine Contractions)')
ax.set_ylabel('(Decelerations Late)')
plt.show()
```

```python
# LB vs MLTV- Might indicate "Mean Value of Long Term Variability" in the FHR.
# From the Graph no credible insight can be drawn
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(newdf['LB'], newdf['MLTV'])
ax.set_xlabel('(Fetal Heart rate)')
ax.set_ylabel('(Mean Value of Long Term Variability)')
plt.show()
```
